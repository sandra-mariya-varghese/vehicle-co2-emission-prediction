```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("co2_emissions (1).csv")
data.head()
```

```python
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
le4=LabelEncoder()
data['make']=le1.fit_transform(data["make"])
data['model']=le2.fit_transform(data["model"])
data['vehicle_class']=le3.fit_transform(data["vehicle_class"])
data['transmission']=le3.fit_transform(data["transmission"])
```

```python
data.head()
```

```python
x=data.drop("co2_emissions",axis=1)
x
```

```python
y=data[["co2_emissions"]]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=123)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in x_train.columns:
    if x_train[col].dtype == 'object':
        x_train[col] = le.fit_transform(x_train[col])
        x_test[col] = le.transform(x_test[col])  # if x_test exists
```

```python
models = {
    'Random Forest': RandomForestRegressor(),
    'DecisionTree Regressor': DecisionTreeRegressor(),
    'GBR': GradientBoostingRegressor(),
    'Linear Regressor': LinearRegression(),
     "SVR": SVR()
}
```

```python
for name, model in models.items():
    model.fit(x_train, y_train)
```

```python
results = {}

for name, model in models.items():
    y_pred = model.predict(x_test)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)  # Root Mean Squared Error
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    results[name] = {
        "MAE": mae,
        "RMSE": rmse,
        "MSE": mse,
        "R²": r2
    }
```

```python
df_results = pd.DataFrame(results).T.sort_values("RMSE")
print(df_results)
```

```python
best_model = models["Random Forest"]  # Based on lowest MSE
best_model
```
